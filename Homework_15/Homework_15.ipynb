```python
# Добавим столбец оценивающий признаки заболевания и прогнозирующий по этим данным состояние лошади
def check_horse_conditions(row):
    """
    1 - норма
    2 - легкая форма заболевания
    3 - тяжелая форма заболевания
    4 - шок
    """
    if row['Температура'] > 38 or row['Тем-ра конеч-ей'] == 2 or row['Боль'] == 4 or row['Вздутие'] == 3 or row['Брюш полость'] == 2:
        return 2
    elif row['Температура'] > 39 or row['Боль'] == 5 or row['Вздутие'] == 4 or row['Брюш полость'] == 3:
        return 3
    elif row['Температура'] < 37.2 or row['Тем-ра конеч-ей'] == 4 or row['Слизистые'] == 3:
        return 4
    else:
        return 1
    
horses['Состояние'] = horses.apply(check_horse_conditions, axis=1)
```

#### Задание 2. Работа с выбросами

В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
def find_ejection(column):
    q1 = horses[column].quantile(0.25)
    q3 = horses[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = horses[horses[column].between(lower_bound, upper_bound, inclusive=True)].sort_values(column)
    result = pd.concat([horses, remove_outliers]).drop_duplicates(keep=False)
    return result.dropna(subset=[column])
```

```python
find_ejection('Пульс')
```

```python
# Повышенная температура корелирует с повышенной частотой дыхания, а также возрастом. Можно предположить, что данные верны и соответствуют молодым особям
# В датафрейме меняем значение возраста 9 на возраст 2, соотвественно описанию.
horses['Возраст'] = horses['Возраст'].replace(9, 2)
```

```python
find_ejection('Частота дыхания')
```

```python
# Для молодых особей оставляем как есть, для взрослых меняем на среднее значение (стандартное отклонение не высокое).
breath_min = find_ejection('Частота дыхания')['Частота дыхания'].min()
breath_mean = horses['Частота дыхания'].mean()
breath_mean
```

```python
horses.loc[(horses['Возраст'] == 1) & (horses['Частота дыхания'] >= breath_min), 'Частота дыхания'] = breath_mean
```

4. Столбец "Тем-ра конеч-ей"

```python
def get_mode(series):
    counts = {}
    for el in series:
        if el not in counts:
            counts[el] = 1
        else:
            counts[el] += 1
    maxw = 0
    mode_ = None
    for key, value in counts.items():
        if maxw < value:
            maxw = value
            mode_ = key
    return mode_
```

```python
horses['Тем-ра конеч-ей'].fillna(horses.groupby('Температура')['Тем-ра конеч-ей'].transform(get_mode), inplace=True)
```

```python
horses[horses['Тем-ра конеч-ей'].isna()]
```
